### Cleaning Data

Memperbaiki tipe data dan nama pada Day Dataset berdasarkan keterangan pada dokumen Readme.txt

```python
#Mengubah keterangan pada column "season"
day["season"] = day["season"].replace({1:"spring", 2:"summer", 3:"fall", 4:"winter"})

#Mengubah keterangan tahun pada column "yr"
day["yr"] = day["yr"].replace({0:"2011", 1:"2012"})

#Mengubah keterangan bulan pada kolom "mnth"
day["mnth"] = day["mnth"].replace({1:"Jan", 2:"Feb", 3:"Mar",4:"Apr",5:"May",6:"Jun",
                                    7:"Jul", 8:"Aug", 9 :"Sep", 10:"Okt", 11:"Nov", 12:"Dec"})

#Mengubah keterangan hari pada column "weekday"
day["weekday"] = day["weekday"].replace({0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'})

#Mengubah keterangan hari kerja pada column "workingday" 
day["workingday"] = day["workingday"].replace({0:"holiday", 1:"working_day"})

#Mengubah keterangan cuaca pada column "weathersit"
day["weathersit"] = day["weathersit"].replace({1:'Clear',2:'Mist',3:'Light Snow',4:'Heavy Rain'})
```

```python
day.head()
```

```python
day["yr"] = pd.to_datetime(day["yr"], format="%Y").dt.year
```

Membuang Column yang tidak digunakan yaitu column "dteday","instant", dan "holiday"

```python
day.drop(["dteday","instant","holiday"], axis=1, inplace=True)
```
